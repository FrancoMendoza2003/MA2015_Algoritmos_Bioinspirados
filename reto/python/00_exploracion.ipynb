```python
import pandas as pd
import numpy as np

from utils.bing import get_location

data_path = "../datos/00_raw/Tsp-1.xlsx"
```

# Procesamiento de datos

```python
data = (
    pd.read_excel(data_path)
    .rename(
        columns={
            "colonia": "colony",
            "Calle": "street",
            "# de casa": "street_number",
            "Código postal": "postal_code",
            "Unnamed: 4": "zone",
        }
    )
    .drop_duplicates()
)
data["zone"] = data["zone"].str.capitalize()
data
```

```python
data["directions"] = (
    data["street"]
    + " "
    + data["street_number"].astype(str)
    + ", "
    + data["colony"]
    + ", "
    + data["zone"]
)
data
```

```python
dirs = data.copy()
dirs[["lat", "lon"]] = dirs.apply(
    lambda row: get_location(
        countryRegion="MX",
        zone=row["zone"],
        locality="Jalisco",
        postalCode=str(row["postal_code"]),
        addressLine=str(row["street_number"]) + " " + row["street"],
        maxResults=1,
    ),
    axis=1,
    result_type="expand",
)

dirs
```

```python
dirs.to_csv(
    "../datos/01_latlon/latlon.csv",
    index=False,
)
```
